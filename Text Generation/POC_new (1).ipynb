```python
from transformers import pipeline
import torch
```

**GPT-neo-125M**

```python
import time
```

```python
#device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
gpt_125 = pipeline('text-generation', model='EleutherAI/gpt-neo-125M',device=0)
```

```python
text="AI/ML is going to play a critical role in solving the problem. Its important to aggregate data of customers in terms of demographics, booking history, service consumption patterns and promotional offers, seasonality etc. We would need the data for AI/ML to comfortably decipher unseen patterns which lead to car cancellations and preempt a possible car cancellation."
```

```python
gpt_125_op,time_record=[],[]
length=[200,190,180,170,160,150,140,130,120,110]
for i in length:
  print("Max_length:",i)
  start=time.time()
  print(gpt_125(text,do_sample=True,max_length=i,min_length=100))
  #gpt_125_op.append(gpt_125(text,do_sample=True,max_length=i,min_length=100))
  end=time.time()
  time_record.append(end-start)
  print("***********************************************************************")
```

```python
import pandas as pd
df=pd.DataFrame(data=gpt_125_op,columns=['output_text'])
df['time_taken']=time_record
df.to_csv("gpt_125.csv",index=False)
```

**Galactica 1.3b**

```python
from transformers import AutoModelForCausalLM,AutoTokenizer
```

```python
pip install accelerate
```

```python
model=AutoModelForCausalLM.from_pretrained("facebook/galactica-1.3b")
tokenizer=AutoTokenizer.from_pretrained('facebook/galactica-1.3b')
```

```python
text=("AI/ML is going to play a critical role in solving the problem. Its important to aggregate data of customers in terms of demographics, booking history, service consumption patterns and promotional offers, seasonality etc. We would need the data for AI/ML to comfortably decipher unseen patterns which lead to car cancellations and preempt a possible car cancellation.")
```

```python
inputs=tokenizer(text,return_tensors='pt').input_ids
output=model.generate(inputs,max_length=200,min_length=100,temperature=0.64,do_sample=True)
print(tokenizer.decode(output[0]))
```

```python
import time
```

```python
galactica_1b_op,time_record=[],[]
inputs=tokenizer(text,return_tensors='pt').input_ids
length=[200,190,180,170,160,150,140,130,120,110]
for i in length:
  print("Max_length:",i)
  start=time.time()
  outputs=model.generate(inputs,max_length=i,min_length=100,do_sample=True)
  #galactica_1b_op.append(tokenizer.decode(outputs[0]))
  print(tokenizer.decode(outputs[0]))
  end=time.time()
  time_record.append(end-start)
  print("Time taken=",(end-start))
  print("***********************************************************************")
```

```python
import pandas as pd
df=pd.DataFrame(data=galactica_1b_op,columns=['output_text'])
df['time_taken']=time_record
df.to_csv("galactica_1b.csv",index=False)
```
